```python
from sympy import *
from sympy.plotting import plot
from sympy.solvers.inequalities import solve_univariate_inequality

init_printing(use_unicode=0, wrap_line=0, no_global=1)
```

```python
x = Symbol("x")
f = x**2 + 3*x - 4
f
```

```python
# Поиск нулей функции
solve(f)
```

```python
f1 = ln(x)
solve(f1)
```

```python
# Поиск неравенства: f > 0 & f < 0
print("f > 0: " + pretty(solve_univariate_inequality(f>0, x)))
print("f < 0: " + pretty(solve_univariate_inequality(f<0, x)))
```

```python
plot(f)
```

```python
from sympy import *
from sympy.plotting import plot

init_printing(use_unicode=0, wrap_line=0, no_global=1)
```

```python
x = Symbol("x")
```

```python
# Симметричная относительно оси о Y и является четной
f = x**2
f
```

```python
plot(f)
```

```python
f.subs(x, 100)
```

```python
f.subs(x, -100)
```

```python
# Не симметрична относительно оси о Y и не является четной
f2 = 2*x**2 + 10*x - 12
f2
```

```python
plot(f2)
```

```python
f2.subs(x, 10)
```

```python
f2.subs(x, -10)
```

```python
f3 = -2*x**2 + 10*x - 12
f3
```

```python
plot(f3)
```

```python
f4 = -2*x**2 + 10*x + 100
f4
```

```python
plot(f4)
```

```python
f5 = sqrt(x)
f5
```

```python
f5.subs(x, -1)
```

```python
plot(f5)
```

```python
plot(f5, (x, 0, 100))
```

```python
f = exp(x)
f
```

```python
f = exp(x) + 1
f
```

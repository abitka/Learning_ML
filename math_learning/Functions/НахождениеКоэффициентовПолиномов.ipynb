```python
# 1. (-2.5, 15), (-2, 10), (-1.5, 7)
"""
a2*x**2 + a1*x + a0

f(-2.5) = 15 -> a2*(-2.5)**2 + a1*(-2.5) + a0 = 15 ->
-> 6.25*a2 - 2.5*a1 + a0 - 15 = 0
f(-2) = 10 -> a2*(-2)**2 + a1*(-2) + a0 = 10 ->
-> 4*a2 - 2*a1 + a0 - 10 = 0
f(-1.5) = 7 -> a2*(-1.5)**2 + a1*(-1.5) + a0 = 7 ->
-> 2.25*a2 - 1.5*a1 + a0 - 7 = 0
"""
eq1_lp = 6.25*a2 - 2.5*a1 + a0 - 15
eq2_lp = 4*a2 - 2*a1 + a0 - 10
eq3_lp = 2.25*a2 - 1.5*a1 + a0 - 7

nonlinsolve([eq1_lp, eq2_lp, eq3_lp], [a2, a1, a0])
```

```python
# 2. (-0.5, -15), (-1, -31), (0, -7) (0.5, -4)
a3, a2, a1, a0 = symbols("a3, a2, a1, a0")
"""
a3*x**3 + a2*x**2 + a1*x + a0

f(-0.5) = -15 -> -0.125*a3 + 0.25*a2 - 0.5*a1 + a0 + 15 = 0
f(-1) = -31 -> -a3 + a2 - a1 + a0 + 31 = 0
f(0) = -7 -> a0 + 7
f(0.5) = -4 -> 0.125*a3 + 0.25*a2 + 0.5*a1 + a0 + 4 = 0
"""
eq1_lp = -0.125*a3 + 0.25*a2 - 0.5*a1 + a0 + 15
eq2_lp = -a3 + a2 - a1 + a0 + 31
eq3_lp = a0 + 7
eq4_lp = 0.125*a3 + 0.25*a2 + 0.5*a1 + a0 + 4

nonlinsolve([eq1_lp, eq2_lp, eq3_lp, eq4_lp], [a3, a2, a1, a0])
```

```python
# 3. (-2, 2), (-1, 1), (0, 0), (1, 1), (2, 2)
# [-2, -1, 0, 1, 2] | [2, 1, 0, 1, 2]
a4, a3, a2, a1, a0 = symbols("a4, a3, a2, a1, a0")
"""
a4*x**4 + a3*x**3 + a2*x**2 + a1*x + a0

f(-2) = 2 -> 16*a4 - 8*a3 + 4*a2 - 2*a1 + a0 - 2
f(-1) = 1 -> a4 - a3 + a2 - a1 + a0 - 1
f(0) = 0 -> a0
f(1) = 1 -> a4 + a3 + a2 + a1 + a0 - 1
f(2) = 2 -> 16*a4 + 8*a3 + 4*a2 + 2*a1 + a0 - 2
"""
eq1_lp = 16*a4 - 8*a3 + 4*a2 - 2*a1 + a0 - 2
eq2_lp = a4 - a3 + a2 - a1 + a0 - 1
eq3_lp = a0
eq4_lp = a4 + a3 + a2 + a1 + a0 - 1
eq5_lp = 16*a4 + 8*a3 + 4*a2 + 2*a1 + a0 - 2

nonlinsolve([eq1_lp, eq2_lp, eq3_lp, eq4_lp, eq5_lp], 
            [a4, a3, a2, a1, a0])
```

```python
# 4. (-3, 33), (-2, 31), (-1, 18), (1, -18), (2, -31), (3, -33)
# [-3, -2, -1, 1, 2, 3] | [33, 31, 18, -18, -31, -33]
a5, a4, a3, a2, a1, a0 = symbols("a5, a4, a3, a2, a1, a0")
"""
a5*x**5 + a4*x**4 + a3*x**3 + a2*x**2 + a1*x + a0

f(-3) = 33 -> (-3)**5*a5 + (-3)**4*a4 + (-3)**3*a3 + (-3)**2*a2 + (-3)*a1 + a0 - 33
f(-2) = 31 -> (-2)**5*a5 + (-2)**4*a4 + (-2)**3*a3 + (-2)**2*a2 - 2*a1 + a0 - 31
f(-1) = 18 -> (-1)**5*a5 + (-1)**4*a4 + (-1)**3*a3 + (-1)**2*a2 - a1 + a0 - 18
f(1) = -18 -> a5 + a4 + a3 + a2 + a1 + a0 + 18
f(2) = -31 -> 2**5*a5 + 2**4*a4 + 2**3*a3 + 2**2*a2 + 2*a1 + a0 + 31
f(3) = -33 -> 3**5*a5 + 3**4*a4 + 3**3*a3 + 3**2*a2 + 3*a1 + a0 + 33
"""
eq1_lp = (-3)**5*a5 + (-3)**4*a4 + (-3)**3*a3 + (-3)**2*a2 + (-3)*a1 + a0 - 33
eq2_lp = (-2)**5*a5 + (-2)**4*a4 + (-2)**3*a3 + (-2)**2*a2 - 2*a1 + a0 - 31
eq3_lp = (-1)**5*a5 + (-1)**4*a4 + (-1)**3*a3 + (-1)**2*a2 - a1 + a0 - 18
eq4_lp = a5 + a4 + a3 + a2 + a1 + a0 + 18
eq5_lp = 2**5*a5 + 2**4*a4 + 2**3*a3 + 2**2*a2 + 2*a1 + a0 + 31
eq6_lp = 3**5*a5 + 3**4*a4 + 3**3*a3 + 3**2*a2 + 3*a1 + a0 + 33

nonlinsolve([eq1_lp, eq2_lp, eq3_lp, eq4_lp, eq5_lp, eq6_lp], 
            [a5, a4, a3, a2, a1, a0])
```

```python
"""
(100, 0), (600, 550)
a1*x + a0
f(100) = 0 -> 100*a1 + a0
f(600) = 550 -> 600*a1 + a0 - 550
"""

eq1_lp = 100*a1 + a0
eq2_lp = 600*a1 + a0 - 550

nonlinsolve([eq1_lp, eq2_lp], [a1, a0])
```

```python
from sympy import *
from sympy.plotting import plot3d

init_printing(use_unicode=0, wrap_line=0, no_global=1)
```

```python
x, y = symbols("x, y")
```

Построить графики нулей

1. f(x, y) = 2x + 5y + 1 <br />
2. f(x, y) = x^2 + y^2 - 1 <br />
Подсказка: для построения графика нулей использейте такой помежуток для переменной в ф-н plot: (y,  -1, 1) <br />
3. f(x, y) = x*cos(y)

```python
# 1. f(x,y) = 2*x + 5*y + 1
f1 = 2*x + 5*y + 1
f1
```

```python
solve(f1)
```

```python
func_f1 = -5/2*y - 1/2
plot(func_f1)
```

```python
plot3d(f1)
```

```python
# 2. f2(x, y) = x**2 + y**2 - 1
f2 = x**2 + y**2 - 1
f2
```

```python
solve(f2)
```

```python
func_f2_1 = -sqrt(1-y**2)
plot(func_f2_1)
```

```python
func_f2_2 = sqrt(1-y**2)
plot(func_f2_2)
```

```python
plot3d(f2)
```

```python
# 3. f3(x, y) = x*cos(y)
f3 = x*cos(y)
f3
```

```python
solve(f3)
```

```python
plot3d(f3)
```

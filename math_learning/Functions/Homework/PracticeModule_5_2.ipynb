Построить графики и найти MSE

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import Symbol
```

```python
# (2, 12.258), (4, 17.24), (8, 30.151)
# f(x) = 3*x + 5
# MSE ~ 1
x = Symbol("x")
points_x = np.array([2, 4, 8])
points_y = np.array([12.258, 17.24, 30.151])
X = np.arange(0, 10, 0.01)
Y = [3*x + 5 for x in X]
plt.scatter(points_x, points_y, c="r")
plt.plot(X, Y, "b")
plt.show()
```

```python
# (2, 3.688), (4, 10.791), (8, 20.705)
# f(x) = 0.25*x**2 + 0.75*x + 1.25
# MSE ~ 4.31
x = Symbol("x")
points_x = np.array([2, 4, 8])
points_y = np.array([3.688, 10.791, 20.705])
X = np.arange(0, 10, 0.01)
Y = [0.25*x**2 + 0.75*x + 1.25 for x in X]
plt.scatter(points_x, points_y, c="r")
plt.plot(X, Y, "b")
plt.show()
```

```python
# (2, 4.872), (4, 29.707), (8, 246.971), (10, 485.727), (12, 840.658)
# f(x) = 0.5*x**3 - 0.25*x**2 + 0.75*x + 1.25
# MSE ~ 3.4
x = Symbol("x")
points_x = np.array([2, 4, 8, 10, 12])
points_y = np.array([4.872, 29.707, 246.971, 485.727, 840.658])
X = np.arange(0, 13, 0.01)
Y = [0.5*x**3 - 0.25*x**2 + 0.75*x + 1.25 for x in X]
plt.scatter(points_x, points_y, c="r")
plt.plot(X, Y, "b")
plt.show()
```

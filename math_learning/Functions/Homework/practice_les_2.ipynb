```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol("x")
```

```python
# 1. f(x) = -3x + 5, where x = 5/3; x = 2
f = -3*x + 5
print(f"1 point: (5/3,{f.subs(x, 5/3)})")
print(f"2 point: (2,{f.subs(x, 2)})")
plot(f, xlim=[-5,5], ylim=[-7,7])
```

```python
# 2. f(x) = 3x^2 + 3x - 6, where x = 0.5; x = 7
f = 3*x**2 + 3*x - 6
print(f"1 point: (0.5,{f.subs(x, 0.5)})")
print(f"2 point: (7,{f.subs(x, 7)})")
plot(f)
```

```python
# 3. f(x) = 2x^4 + x^3 - 6x^2 + 7x + 1, where x = 3; x = -5
f = 2*x**4 + x**3 - 6*x**2 + 7*x + 1
print(f"1 point: (3,{f.subs(x, 3)})")
print(f"2 point: (-5,{f.subs(x, -5)})")
plot(f)
```

```python
# 4. f(x) = (x^2)^x, where x = 0; x = 10
f = (x**2)**x
print(f"1 point: (0,{f.subs(x, 0)})")
print(f"2 point: (10,{f.subs(x, 10)})")
plot(f)
```

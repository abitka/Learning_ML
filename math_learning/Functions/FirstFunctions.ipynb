```python
from sympy import *
from sympy.plotting import plot

init_printing(use_unicode=0, wrap_line=0, no_global=1)
```

```python
x = Symbol("x")

f = x + 1
f
```

```python
# передаем значение в x
print(f"1 point: (0,{f.subs(x, 0)})")
```

```python
print(f"2 point: (-1,{f.subs(x, -1)})")
```

```python
# строим график функции f(x) = x + 1
plot(f, xlim=[-3,3], ylim=[-3,3])
```

```python
from sympy import *
from sympy.plotting import plot3d

init_printing(use_unicode=0, wrap_line=0, no_global=1)
```

```python
x, y = symbols("x, y")

f = 5*x - 3*y + 2
f
```

```python
solve(f)
```

```python
func_x = 3/5*y - 2/5
func_x
```

```python
plot(func_x)
```

```python
plot3d(f)
```

```python
f = x*y / sqrt(x**2 + y**2 - 5)
f
```

```python
solve(f)
```
